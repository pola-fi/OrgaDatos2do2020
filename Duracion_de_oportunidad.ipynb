```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from matplotlib import pyplot as plt

#Levantamos el archivo que ya tiene los datos principales depurados
df = pd.read_csv("./data/train_clean.csv")

#Normalizo Fechas ya que las usar para el analisis
df['Account_Created_Date'] = pd.to_datetime(df['Account_Created_Date'], format='%m/%d/%Y')
df['Last_Modified_Date'] = pd.to_datetime(df['Last_Modified_Date'], format='%m/%d/%Y')
df['Quote_Expiry_Date'] = pd.to_datetime(df['Quote_Expiry_Date'], format='%m/%d/%Y')
df['Planned_Delivery_Start_Date'] = pd.to_datetime(df['Planned_Delivery_Start_Date'], format='%m/%d/%Y')
df['Planned_Delivery_End_Date'] = pd.to_datetime(df['Planned_Delivery_End_Date'], format='%m/%d/%Y')
df['Account_Created_Date'] = pd.to_datetime(df['Account_Created_Date'], format='%m/%d/%Y')
df['Opportunity_Created_Date'] = pd.to_datetime(df['Opportunity_Created_Date'], format='%m/%d/%Y')
df['Total_days_until_delivery'] = (df['Planned_Delivery_End_Date']) - (df['Opportunity_Created_Date'])

#Tomo los el rango de das entre la fecha de creacin de la oportunidad y la ltima fecha de entrega y los agrupo por rangos
#redondeando para arriba, para luego contar "Closed Won" y "Closed Lost" segn rangos
#Para los pocos valores menores a 0 (los que la fecha de entrega era anterior a la fecha de la creacin de la oportunidad),
#puedo suponer que eran pedidos por encargo o urgentes, y tomo la decisin de agruparlos en el rango de 0 das para simplificar el anlisis
#Lo mismo para las oportunidades que fueron entregadas despus de 350 das, las agrupo en el rango de 360 das
df.loc[df['Total_days_until_delivery'] <= timedelta(0),'Total_days_until_delivery'] = timedelta(0)
df.loc[(timedelta(0) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(10)),'Total_days_until_delivery'] = timedelta(10)
df.loc[(timedelta(10) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(20)),'Total_days_until_delivery'] = timedelta(20)
df.loc[(timedelta(20) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(30)),'Total_days_until_delivery'] = timedelta(30)
df.loc[(timedelta(30) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(40)),'Total_days_until_delivery'] = timedelta(40)
df.loc[(timedelta(40) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(50)),'Total_days_until_delivery'] = timedelta(50)
df.loc[(timedelta(50) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(60)),'Total_days_until_delivery'] = timedelta(60)
df.loc[(timedelta(60) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(70)),'Total_days_until_delivery'] = timedelta(70)
df.loc[(timedelta(70) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(80)),'Total_days_until_delivery'] = timedelta(80)
df.loc[(timedelta(80) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(90)),'Total_days_until_delivery'] = timedelta(90)
df.loc[(timedelta(90) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(100)),'Total_days_until_delivery'] = timedelta(100)
df.loc[(timedelta(100) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(110)),'Total_days_until_delivery'] = timedelta(110)
df.loc[(timedelta(110) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(120)),'Total_days_until_delivery'] = timedelta(120)
df.loc[(timedelta(120) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(130)),'Total_days_until_delivery'] = timedelta(130)
df.loc[(timedelta(130) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(140)),'Total_days_until_delivery'] = timedelta(140)
df.loc[(timedelta(140) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(150)),'Total_days_until_delivery'] = timedelta(150)
df.loc[(timedelta(150) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(160)),'Total_days_until_delivery'] = timedelta(160)
df.loc[(timedelta(160) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(170)),'Total_days_until_delivery'] = timedelta(170)
df.loc[(timedelta(170) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(180)),'Total_days_until_delivery'] = timedelta(180)
df.loc[(timedelta(180) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(190)),'Total_days_until_delivery'] = timedelta(190)
df.loc[(timedelta(190) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(200)),'Total_days_until_delivery'] = timedelta(200)
df.loc[(timedelta(200) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(210)),'Total_days_until_delivery'] = timedelta(210)
df.loc[(timedelta(210) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(220)),'Total_days_until_delivery'] = timedelta(220)
df.loc[(timedelta(220) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(230)),'Total_days_until_delivery'] = timedelta(230)
df.loc[(timedelta(230) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(240)),'Total_days_until_delivery'] = timedelta(240)
df.loc[(timedelta(240) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(250)),'Total_days_until_delivery'] = timedelta(250)
df.loc[(timedelta(250) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(260)),'Total_days_until_delivery'] = timedelta(260)
df.loc[(timedelta(260) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(270)),'Total_days_until_delivery'] = timedelta(270)
df.loc[(timedelta(270) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(280)),'Total_days_until_delivery'] = timedelta(280)
df.loc[(timedelta(280) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(290)),'Total_days_until_delivery'] = timedelta(290)
df.loc[(timedelta(290) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(300)),'Total_days_until_delivery'] = timedelta(300)
df.loc[(timedelta(300) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(310)),'Total_days_until_delivery'] = timedelta(310)
df.loc[(timedelta(310) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(320)),'Total_days_until_delivery'] = timedelta(320)
df.loc[(timedelta(320) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(330)),'Total_days_until_delivery'] = timedelta(330)
df.loc[(timedelta(330) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(340)),'Total_days_until_delivery'] = timedelta(340)
df.loc[(timedelta(340) < df['Total_days_until_delivery']) & (df['Total_days_until_delivery'] <= timedelta(350)),'Total_days_until_delivery'] = timedelta(350)
df.loc[df['Total_days_until_delivery'] > timedelta(350),'Total_days_until_delivery'] = timedelta(360)

#Creo 2 columnas para saber si se obtuvo Closed Won o Closed Lost segn si vale 1 o 0
df['Closed_Lost'] = 0
df['Closed_Won'] = 0
df.loc[df['Stage'] == 'Closed Lost', 'Closed_Lost'] = 1
df.loc[df['Stage'] == 'Closed Won', 'Closed_Won'] = 1

#Filtramos las columnas que nos interesan y creamos una columna con la probabilidad de que suceda Closed Won segn el grupo
df_filtered = df[['Total_days_until_delivery', 'Closed_Won', 'Closed_Lost']]
df_filtered = df_filtered.groupby('Total_days_until_delivery').sum()
df_filtered['Probability_to_closed_won'] = df_filtered['Closed_Won'] / (df_filtered['Closed_Won'] + df_filtered['Closed_Lost'])
df_filtered['Range_of_days_until_delivery'] = df_filtered.index.days

x = df_filtered['Range_of_days_until_delivery']
y = df_filtered['Probability_to_closed_won']

plt.fill_between( x, y, color="skyblue", alpha=0.2)
plt.rcParams["figure.figsize"] = [20, 12]
plt.title("Probabilidad de venta concretada en funcin de los das")
plt.xlabel("Das entre oportunidad creada y ltimo da de entrega")
plt.ylabel("Probabilidad de venta concretada")

plt.plot(x, y, color="Slateblue", alpha=1)
```
