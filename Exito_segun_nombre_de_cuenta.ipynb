```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from matplotlib import pyplot as plt

#Levantamos el archivo que ya tiene los datos principales depurados
df = pd.read_csv("./data/train_clean.csv")
#df.info()
# #Normalizo Floats
df['Total_Amount'] = pd.to_numeric(df['Total_Amount'],downcast='float',errors='coerce')
df['Total_Taxable_Amount'] = pd.to_numeric(df['Total_Taxable_Amount'],downcast='float',errors='coerce')
df['Total_Taxable_Amount_Currency'] = df['Total_Taxable_Amount_Currency'].astype(str)
#df.info()
#Creo 2 columnas para saber si se obtuvo Closed Won o Closed Lost según si vale 1 o 0
df['Closed_Lost'] = 0
df['Closed_Won'] = 0
df.loc[df['Stage'] == 'Closed Lost', 'Closed_Lost'] = 1
df.loc[df['Stage'] == 'Closed Won', 'Closed_Won'] = 1
#df.loc[df['Stage'] == 'Closed Won', 'TRF_sold'] = df['TRF']
df['Total_oportunities'] = df['Closed_Won'] + df['Closed_Lost']
#Conversion de lo facturado a dólares
#df['Total_Taxable_Amount_usd'] = df['Total_Taxable_Amount']
# df.loc[df['Total_Taxable_Amount_Currency'] == 'JPY', 'Total_Taxable_Amount_usd'] = df['Total_Taxable_Amount_usd']/112.5
# df.loc[df['Total_Taxable_Amount_Currency'] == 'EUR', 'Total_Taxable_Amount_usd'] = df['Total_Taxable_Amount_usd']*1.1427 
# df.loc[df['Total_Taxable_Amount_Currency'] == 'GPB', 'Total_Taxable_Amount_usd'] = df['Total_Taxable_Amount_usd']*1,34092
# df.loc[df['Total_Taxable_Amount_Currency'] == 'AUD', 'Total_Taxable_Amount_usd'] = df['Total_Taxable_Amount_usd']*0,7642
#df.loc[df['Stage'] == 'Closed Won', 'Total_Taxable_Amount_in_Won'] = df['Total_Taxable_Amount']
df_filtered = df.groupby('Account_Name').sum()
df_filtered['Probability_to_closed_won'] = df_filtered['Closed_Won'] / (df_filtered['Closed_Won'] + df_filtered['Closed_Lost'])
df_filtered = df_filtered.loc[df_filtered['Total_oportunities'] > 50]

df_filtered['Mean_TRF_per_opportunity'] = df_filtered['TRF'] / df_filtered['Total_oportunities']
df_filtered.sort_values(by='Probability_to_closed_won', ascending=False, inplace=True)
df_filtered.head(50)
```
